```python
import json
import requests
from bs4 import BeautifulSoup
import bs4
import pandas as pd
```

```python
for i in range(30,105):
        x="https://www.bkmkitap.com/bkmkitapfuari?view=0&pg="+str(i)
        def get_page_contents(x):
            page = requests.get(x)
            return bs4.BeautifulSoup(page.text, "html.parser")

        soup = get_page_contents(x)
        for linkler2 in soup.find_all('a', {"class":"image-wrapper"}):
            a = (linkler2.get("href"))
            url = "https://www.bkmkitap.com"+str(a)
            
            soup2 = get_page_contents(url)
   
            Product_Name = soup2.find('h1', {"id":"productName"})
            Publisher = soup2.find('a', {"id":"product-brand"})
            Writer = soup2.find('a', {"id":"productModelText"})
            Gain = soup2.find('span', {"id":"gain"})
            Discount = soup2.find('span', {"class":"productDiscount"})
            Current_Price = soup2.find('span', {"class":"product-price"})
            Previous_Price = soup2.find('span', 
                                {"class":"product-price-not-discounted"})
            i1 = Product_Name.text
            i2 = Publisher.text.strip()
            i3 = Writer.text.strip()
            i4 = Previous_Price.text
            i5 = Discount.text.strip()
            i6 = Current_Price.text
            i7 = Gain.text
            
            NameList.append(i1)
            PublisherList.append(i2)
            WriterList.append(i3)
            Previous_PriceList.append(i4)
            DiscountList.append(i5)
            Current_PriceList.append(i6)
            GainList.append(i7)
            tr = soup2.find('div', {'class':'cilt'})
            KindList.append(tr.find_all('span')[1].text)
            PageList.append(tr.find_all('span')[3].text)
            ISBNcodeList.append(tr.find_all('span')[5].text)
            CoverList.append(tr.find_all('span')[7].text)
                        
            bkmkitaplar8 = pd.DataFrame({
                'Product_Name': NameList, 
                'Publisher': PublisherList,
                'Writer': WriterList, 
                'Previous_Price': Previous_PriceList, 
                'Discount': DiscountList,
                'Current_Price': Current_PriceList,
                "Gain": GainList,
                "Kind": KindList,
                "Page": PageList,
                "ISBN": ISBNcodeList,
                "Cover": CoverList
            })
bkmkitaplar8
           
```
